```python
from ckanapi import RemoteCKAN
import json
```

```python
dataset = api.action.package_show(id = DATASET_ID)
print(dataset)
```

```python
for resources in dataset["resources"]:
    print(f"Nom : {resources['name']}")
    print(f"format : {resources['format']}")
```
